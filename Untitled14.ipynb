```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('train.csv')
```

```python
# Display basic information about the dataset
print(df.info())

```

```python
# Display summary statistics of numerical columns
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Explore data distributions using histograms
df.hist(bins=50, figsize=(10, 10))
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame
# Adjust the figsize to make room for each subplot
df.hist(bins=50, figsize=(20, 20))

# Adjust the layout
plt.tight_layout()

# Optional: Add a grid for better readability
plt.grid(True)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np  # Ensure NumPy is imported

# Select only numeric columns for the correlation matrix
numeric_df = df.select_dtypes(include=[np.number])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)

# Show plot
plt.show()

```

```python
# Explore categorical variables using count plots
categorical_columns = df.select_dtypes(include=['object']).columns
for column in categorical_columns:
    plt.figure(figsize=(8, 10))
    sns.countplot(x=column, data=df)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
data = pd.read_csv('train.csv')  # Replace 'path_to_your_file.csv' with your file path

# Plot histograms for all numerical columns
def plot_numerical_histograms(data):
    numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
    for col in numerical_cols:
        plt.figure(figsize=(10, 5))
        sns.histplot(data[col], kde=True)
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.show()
plot_numerical_histograms(data)
```

```python
# Plot histograms for all categorical columns
def plot_categorical_histograms(data):
    categorical_cols = data.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        plt.figure(figsize=(10, 5))
        sns.countplot(y=data[col], order=data[col].value_counts().index)
        plt.title(f'Count of {col}')
        plt.xlabel('Count')
        plt.ylabel(col)
        plt.show()
plot_categorical_histograms(data)
```

```python
def plot_correlation_heatmap(data):
    numerical_data = data.select_dtypes(include=['float64', 'int64'])
    correlation_matrix = numerical_data.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Heatmap of Numerical Data Correlation')
    plt.show()
plot_correlation_heatmap(data)
```
